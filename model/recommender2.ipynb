## Trying out KNN

```python
df_features = df.drop(columns=['ArtistName', 'SongName', 'Name'])
```

```python
df_features
```

```python
scaler = StandardScaler()
normalized_data = scaler.fit_transform(df_features)
normalized_data 
```

```python
knn = NearestNeighbors(n_neighbors=13, algorithm='auto', metric='cosine')  
```

```python
knn.fit(normalized_data)
```

```python
distances, indices = knn.kneighbors([normalized_data[1]])
distances, indices
```

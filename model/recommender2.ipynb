## Trying out KNN

```python
df_features = df.drop(columns=['ArtistName', 'SongName', 'Name'])
```

```python
df_features
```

```python
scaler = StandardScaler()
normalized_data = scaler.fit_transform(df_features)
normalized_data 
```

```python
knn = NearestNeighbors(n_neighbors=13, algorithm='auto', metric='cosine')  
```

```python
knn.fit(normalized_data)
```

```python
distances, indices = knn.kneighbors([normalized_data[1]])
distances, indices
```

```python
indices
```

```python
df.shape , df_features.shape
```

```python
df_features['S.No'] = range(1, len(df_features) + 1)
```

```python
df['S.No'] = range(1, len(df) + 1)
```

```python
df
```

```python
for i in range(10):
    song_id = i
    distances, indices = knn.kneighbors([normalized_data[song_id]])
    # print(indices)
    print(f"Songs recommended for {df['Name'][song_id]} : \n")
    indices = indices[:2]
    for index in indices:
        print(df['Name'][index])
```

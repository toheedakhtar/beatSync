```python
def give_recommendations(index=None, songName = None):
    try:
        if songName is not None:
            index = songs_df_cpy[songs_df_cpy['SongName'] == songName]['id'].values[0]
        index_recomm = cos_sim_data.loc[index].sort_values(ascending=False).index.tolist()[1:5]
    
        name = songs_df_cpy['SongName'].iloc[index] if songName is None else songName
        print(f"Recommending for : {name}")
        for index_r in index_recomm:
            songs_recomm = songs_df_cpy['SongName'].iloc[index_r]
            print(songs_recomm)
        # for song in songs_recomm:
        #     print(f"{song} is recommended for {songs_df_cpy['SongName'].loc[index]}")
    except:
        print(index)
```
